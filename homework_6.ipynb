```python
import pandas as pd

csv_file_path = "jamb_exam_results.csv"

df = pd.read_csv(csv_file_path) 

print(df.head())
```

```python
df
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df
```

```python
del df['student_id']
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.fillna(0)
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=1)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=1)

len(df_train), len(df_val), len(df_test)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

y_train = df_train.jamb_score.values
y_val = df_val.jamb_score.values
y_test = df_test.jamb_score.values

del df_train['jamb_score']
del df_val['jamb_score']
del df_test['jamb_score']
```

```python
from sklearn.feature_extraction import DictVectorizer

dv = DictVectorizer(sparse=False)

train_dicts = df_train[df_train.columns].to_dict(orient='records')
X_train = dv.fit_transform(train_dicts)

val_dicts = df_val[df_val.columns].to_dict(orient='records')
X_val = dv.fit_transform(val_dicts)

test_dicts = df_test[df_test.columns].to_dict(orient='records')
X_test = dv.fit_transform(test_dicts)
```

**Question 1**

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(max_depth=1)
dt.fit(X_train, y_train)
```

```python
index = dt.tree_.feature[0]
feature = dv.feature_names_[index]
feature
```

```python
from sklearn.tree import export_text
print(export_text(dt, feature_names=list(dv.get_feature_names_out())))
```

**Question 2**

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
dt = RandomForestRegressor(n_estimators=10, random_state=1, n_jobs=-1)
dt.fit(X_train, y_train)
```

```python
from sklearn.metrics import root_mean_squared_error

y_pred = dt.predict(X_val)
rsme = root_mean_squared_error(y_val, y_pred)
rsme
```

**Question 3**

```python
scores = []

for n in range(10, 201, 10):
    dt = RandomForestRegressor(n_estimators=n, random_state=1, n_jobs=-1)
    dt.fit(X_train, y_train)
    y_pred = dt.predict(X_val)
    rsme = root_mean_squared_error(y_val, y_pred)
    scores.append((n, round(rsme, 3)))
```

```python
scores
```

```python
import matplotlib.pyplot as plt

df_scores = pd.DataFrame(scores, columns=['n_estimators', 'rsme'])
plt.plot(df_scores.n_estimators, df_scores.rsme)
```

80 is the nearest answer

```python
max_depth = [10, 15, 20, 25]
scores= []
for m in max_depth:
    for n in range(10, 201, 10):
        dt = RandomForestRegressor(n_estimators=n, max_depth=m , random_state=1, n_jobs=-1)
        dt.fit(X_train, y_train)
        y_pred = dt.predict(X_val)
        rsme = root_mean_squared_error(y_val, y_pred)
        scores.append((m, n, round(rsme, 3)))
```

```python
columns = ['max_depth', 'n_estimators', 'rsme']
df_scores = pd.DataFrame(scores, columns=columns)
df_scores
```

```python
df_scores_pivot = df_scores.pivot(index='n_estimators', columns=['max_depth'], values=['rsme'])
df_scores_pivot
```

```python
import seaborn as sns

sns.heatmap(df_scores_pivot, annot=True, fmt=".3f")
```

max_depth 10 has the lowest Rsme

**Question 6**

```python
import xgboost as xgb
```

```python
features = list(dv.get_feature_names_out())
dtrain = xgb.DMatrix(X_train, label=y_train, feature_names=features)
dval = xgb.DMatrix(X_val, label=y_val, feature_names=features)
```

```python
watchlist = [(dtrain, 'train'), (dval, 'val')]
```

```python
xgb_params = {
    'eta': 0.3, 
    'max_depth': 6,
    'min_child_weight': 1,
    
    'objective': 'reg:squarederror',
    'nthread': 8,
    
    'seed': 1,
    'verbosity': 1,
}

model = xgb.train(xgb_params, dtrain, num_boost_round=100, evals=watchlist)
```

The lowest rmse for eta = 0.3 : [7]	train-rmse: 33.07449	val-rmse: 40.65074

```python
xgb_params = {
    'eta': 0.1, 
    'max_depth': 6,
    'min_child_weight': 1,
    
    'objective': 'reg:squarederror',
    'nthread': 8,
    
    'seed': 1,
    'verbosity': 1,
}

model = xgb.train(xgb_params, dtrain, num_boost_round=100, evals=watchlist)
```

the lowest rmse for eta = 0.1 : [38]	train-rmse: 30.15893	val-rmse: 40.11076

So eta = 0.1 leads to the best rmse score
